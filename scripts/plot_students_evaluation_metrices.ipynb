```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

```python
#load the csv file generated from the SQL query
df=pd.read_csv('../outputs/csvs/student_evaluation.csv')
df.head()
```

```python
#select the numeric columns
df_numeric=df.select_dtypes(np.number)
print(df_numeric.columns)
```

```python
#pairplot
fig=plt.Figure(figsize=(16, 16))
sns.pairplot(data=df_numeric)
plt.savefig('../outputs/figures/students_evaluation_pairplots.png', dpi=300)
```

```python
#heatmap for the correlations
corr=df_numeric.corr()
fig=plt.Figure(figsize=(10,15))
ax=sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.tight_layout()
plt.savefig('../outputs/figures/students_evaluation_correlation_heatmap.png', dpi=300, bbox_inches='tight')
```
